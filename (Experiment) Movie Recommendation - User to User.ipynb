## Create the procedure

```python
pivot_df = pd.pivot_table(df, values='rating', index='userId', columns='movieId')
```

```python
pivot_df
```

```python
# move the matrix to 0 center
# handled "tough rater" and "easy rater"
avg_rating_array = pivot_df.mean(axis=1)
norm_pivot_df = pivot_df.sub(avg_rating_array, axis='rows')
norm_pivot_df.fillna(0, inplace=True)
```

```python
norm_pivot_df
```

```python
# calculate Centered Cosine Similirity (Pearson Correlation)
user_similarity_df = norm_pivot_df.T.corr(method='pearson')
```

```python
user_similarity_df
```

```python
# user 1 movie 2
(pivot_df.loc[:, 2] * user_similarity_df.loc[:, 1]).sum() / user_similarity_df.loc[:, 1].sum()
```

```python
def predict_rating(user_id, movie_id):
    """Return a rating number by calculate the weighted average of similar users rating"""
    nom = (pivot_df.loc[:, movie_id] * user_similarity_df.loc[:, user_id]).sum()
    denom = user_similarity_df.loc[:, user_id].sum()
    return nom / denom
```

```python
def recommend_movies(user_id, rating_threshold=3):
    """Return a dictionary of recommend movie_id:predict_rating if the predict_rating larger than threshold"""
    movie_ids = pivot_df.columns
    recommend_movies = {}
    for movie in movie_ids:
        prediction = predict_rating(user_id, movie)
        if prediction > rating_threshold:
            recommend_movies[movie] = prediction
    return recommend_movies
```

```python
movie_df = pd.read_csv("ml-latest-small/movies.csv")
```

```python
movie_df.head()
```

```python
def print_recommendation(user_id):
    watched_movie_ids = pivot_df.T.loc[:, user_id][pivot_df.T.loc[:, user_id].notnull()].index
    favorite_movie_ids = pivot_df.T.loc[:, user_id].sort_values(ascending=False).head(5).index
    print(f"User {user_id}'s favorite movies: ")
    for movie_id in favorite_movie_ids:
        print(movie_df.loc[movie_df['movieId']==movie_id, 'title'].values, '\n')

    recommend_movies_dict = recommend_movies(user_id, rating_threshold=2.5)
    print(f"User {user_id}'s recomend movies: ")
    for k, v in recommend_movies_dict.items():
        if k not in watched_movie_ids:
            print(movie_df.loc[movie_df['movieId']==k, 'title'].values, v, '\n')
```
