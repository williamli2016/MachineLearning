## Load Data

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data=data.data, columns=data.feature_names)
```

```python
df['target'] = data.target
```

## Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score, log_loss, classification_report
```

```python
# train, test split
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.3,
                                                                random_state=5)
```

```python
cls = DecisionTreeClassifier(min_samples_leaf=5)
```

```python
cls.fit(X_train, y_train)
```

```python
predictions = cls.predict(X_validation)
```

```python
# evaluate
print("Model AUC score: ", roc_auc_score(y_validation, predictions))
print("Model log_loss score: ", log_loss(y_validation, predictions))
```

```python
print(classification_report(y_validation, predictions, target_names=data.target_names))
```

## Visualize Decision Tree

```python
from sklearn import tree
import graphviz
```

```python
fig, ax = plt.subplots(figsize=(20, 15))
tree.plot_tree(cls, filled=True, feature_names=data.feature_names,
               class_names=data.target_names, fontsize=10)
plt.show()
```
