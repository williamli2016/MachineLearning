## Load Data

```python
user_ratings = pd.merge(ratings[['userId', 'movieId', 'rating']], 
                        movies[['movieId', 'title']],
                        on='movieId')
```

## Manipulate Data

```python
pivot_user_ratings = pd.pivot_table(user_ratings,
                                    index='userId',
                                    columns='title',
                                    values='rating')
```

```python
pivot_user_ratings.head()
```

We want to avoid spurious results that happened from just a handful of users that happened to rate the same pair of movies. In order to retrict our results to movies that lots of people rated together - and also give us more popular results that are more easily recongnizable - we'll use the min_periods argument to throw out results where fewer than 100 users rated a given movie pair.

```python
corr_matrix = pivot_user_ratings.corr(method='pearson', min_periods=100)
```

```python
corr_matrix.head()
```

Let's produce some movie recommendations for user ID 1

```python
my_ratings = pivot_user_ratings.loc[100].dropna()
my_ratings
```

Let's go through each movie I rated one at a time, and build up a list of possible recommendations based on the movies similar to the ones I rated.

```python
similar_candidates = pd.Series()
for i in range(0, len(my_ratings.index)):
    print(f"Adding sims for {my_ratings.index[i]}...")
    # Retrieve similar movies to this one that I rated
    sims = corr_matrix[my_ratings.index[i]].dropna()
    # Now scale its similarity by how well I rated this movie
    sims = sims.map(lambda x: x * my_ratings[i])
    # Add the score to the list of similarity candidates
    similar_candidates = similar_candidates.append(sims)

# Glance at our results so far:
print("Sorting...")
similar_candidates.sort_values(inplace=True, ascending=False)
print(similar_candidates.head(10))
```

Note that some of the same movies came up more than one, because they were similar to more than one movie I rated.We'll use groupby() to add together the scores from movies that show up more than once.

```python
similar_candidates = similar_candidates.groupby(similar_candidates.index).sum()
```

```python
similar_candidates.sort_values(inplace=True, ascending=False)
```

```python
similar_candidates.head(10)
```

Last thing we have to do is filter out movies I've already rated.

```python
filtered_similar = similar_candidates.drop(my_ratings.index, errors='ignore')
```
