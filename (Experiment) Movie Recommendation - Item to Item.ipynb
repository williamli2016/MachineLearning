## Create the procedure

```python
pivot_df = pd.pivot_table(df, values='rating', index='movieId', columns='userId')
```

```python
pivot_df
```

```python
# move the matrix to 0 center
avg_rating_array = pivot_df.mean(axis=1)
norm_pivot_df = pivot_df.sub(avg_rating_array, axis='rows')
norm_pivot_df.fillna(0, inplace=True)
```

```python
norm_pivot_df
```

```python
# calculate Centered Cosine Similirity (Pearson Correlation)
movie_similarity_df = norm_pivot_df.T.corr(method='pearson')
```

```python
movie_similarity_df
```

```python
def predict_rating(user_id, movie_id):
    """Return a rating number by calculate the weighted average of similar movies rating"""
    corr_lt_0_index = movie_similarity_df.loc[:, movie_id][movie_similarity_df.loc[:, movie_id] > 0].index  ##select only the neighborhood with corr > 0
    nom = (pivot_df.loc[:, user_id][corr_lt_0_index] * movie_similarity_df.loc[:, movie_id][corr_lt_0_index]).sum()
    denom = movie_similarity_df.loc[:, movie_id].sum()
    return nom / denom
```

```python
def recommend_movies(user_id, rating_threshold=3):
    """Return a dictionary of recommend movie_id:predict_rating if the predict_rating larger than threshold"""
    movie_ids = pivot_df.index
    recommend_movies = {}
    for movie in movie_ids:
        prediction = predict_rating(user_id, movie)
        if prediction > rating_threshold:
            recommend_movies[movie] = prediction
    return recommend_movies
```

```python
movie_df = pd.read_csv("Datasets/ml-latest-small/movies.csv")
```

```python
def print_recommendation(user_id):
    watched_movie_ids = pivot_df.loc[:, user_id][pivot_df.loc[:, user_id].notnull()].index
    favorite_movie_ids = pivot_df.loc[:, user_id].sort_values(ascending=False).head(5).index
    print(f"User {user_id}'s favorite movies: ")
    for movie_id in favorite_movie_ids:
        print(movie_df.loc[movie_df['movieId']==movie_id, ['title', 'genres']].values, '\n')

    recommend_movies_dict = recommend_movies(user_id, rating_threshold=4)
    print(f"User {user_id}'s recomend movies: ")
    for k, v in recommend_movies_dict.items():
        if k not in watched_movie_ids:
            print(movie_df.loc[movie_df['movieId']==k, ['title', 'genres']].values, v, '\n')
```
